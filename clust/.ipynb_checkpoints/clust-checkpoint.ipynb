```python
def show_dataframe(df: pd.DataFrame):
    display(HTML(df.to_html()))
```

```python
def dataset(csv_url, class_column_name='class'):
    df = pd.read_csv(csv_url)
    show_dataframe(df)
    # y, _ = pd.factorize(ds[class_column_name])
```
